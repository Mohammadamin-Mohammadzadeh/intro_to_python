```python
with open('words.txt', mode='r') as file:
    words = file.readlines()
words
```

```python
clean_words = sorted(set([w.strip().lower() for w in words]))
clean_words
```

```python
def get_signiture(word):
    return ''.join(sorted(word))
```

```python
get_signiture('elvis')
```

```python
word = 'dictionary'
get_signiture(word)
```

```python
[w for w in clean_words if get_signiture(w) == get_signiture('dictionary')]
```

```python
%timeit [w for w in clean_words if get_signiture(w) == get_signiture('dictionary')]
```
